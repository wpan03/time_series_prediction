```python
import yfinance as yf
```

```python
def process_yf_data(ticker: yf.ticker.Ticker, symbol_name: str):
    df = ticker.history(period="3mo").reset_index()
    df.columns = map(str.lower, df.columns)
    drop_col = ['dividends', 'stock splits', 'capital gains']
    df = df.drop(drop_col, axis=1)
    df.columns = ['date'] + [f'{symbol_name}_{col}' for col in df.columns if col != 'date']
    return df
```

```python
spy = yf.Ticker("SPY")
qqq = yf.Ticker("QQQ")

df_prepare = process_yf_data(spy, 'spy').merge(process_yf_data(qqq, 'qqq'), on='date')
```
